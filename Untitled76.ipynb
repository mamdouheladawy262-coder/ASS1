```python
import numpy as np

# Set random seed (optional, for reproducibility)
np.random.seed(0)

# Input (example)
x = np.array([[0.6],
              [0.2]])   # 2x1 input vector

# Random weights in [-0.5, 0.5]
W1 = np.random.uniform(-0.5, 0.5, (2, 2))  # Hidden layer weights (2 neurons)
W2 = np.random.uniform(-0.5, 0.5, (1, 2))  # Output layer weights

# Biases
b1 = 0.5
b2 = 0.7

# Tanh activation function
def tanh(x):
    return np.tanh(x)

# Forward propagation
z1 = np.dot(W1, x) + b1
a1 = tanh(z1)

z2 = np.dot(W2, a1) + b2
output = tanh(z2)

# Print result
print("Network output:")
print(output)
```
